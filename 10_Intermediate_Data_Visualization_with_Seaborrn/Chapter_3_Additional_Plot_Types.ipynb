In the first two chapters of this course, we covered the basics of how to use the Seaborn API for creating and customizing plots using different Seaborn and matplotlib approaches. This lesson will focus on the many different categorical plots that Seaborn supports.

### Categorical Data

In our earlier exercises we looked at distrubition and linear regression plots, used on numerical values. Seaborn also supports many plot types with categorical data.

**Categorical Data**:

* Data which takes on a limited and fixed number of values 
* Normally combined with numeric data
* Examples include:
    - Geography (country, state, region)
    - Gender
    - Ethnicity
    - Blood type
    - Eye color

In this lesson, we will bne looking at US Healthcare reimbursement values. In these examples, the codes are the categorical variables and the average hospital charge is the numerical value we will analyze in our plots.

## Exercise I

### Stripplot() and swarmplot()

Many datasets have categorical data and Seaborn supports several useful plot types for this data. 
In this example, we will continue to look at the 2010 School Improvement data and segment the data 
by the types of school improvement models used.

````python
# Import modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset
df = pd.read_csv('datasets/schoolimprovement2010grants.csv')
````

````python
# Create the stripplot
sns.stripplot(data=df, x="Award_Amount",
              y="Model Selected",
              jitter=True)

plt.show()
````

````python
# Create and display a swarmplot with hue set to the Region
sns.swarmplot(data=df, x="Award_Amount",
              y="Model Selected",
              hue="Region")

plt.show()
````

### boxplots, violinplots and boxenplots

Seaborn's categorical plots also support several abstract representations of data. The API for each 
of these is the same so it is very convenient to try each plot and see if the data lends itself to 
one over the other.

In this exercise, we will use the color palette options presented in Chapter 2 to show how colors 
can easily be included in the plots.

````python
# Create a boxplot
sns.boxplot(data=df, x="Award_Amount",
            y="Model Selected")

plt.show()
plt.clf()
````

````python
# Create a violinplot with the husl palette
sns.violinplot(data=df, x="Award_Amount",
               y="Model Selected",
               palette='husl')

plt.show()
plt.clf()
````

````python
# Create a boxenplot with the Paired palette and the Region column as the hue
sns.boxenplot(data=df, x="Award_Amount",
              y="Model Selected",
              palette="Paired",
              hue="Region")

plt.show()
plt.clf()
````

### barplots, pointplots and countplots

The final group of categorical plots are ```barplots```, ```pointplots``` and ```countplot``` which create statistical summaries of the data. The plots follow a similar API as the other plots and allow further customization for the specific problem at hand.

````python
# Show a countplot with the number of models used with each region a different color
sns.countplot(data=df, y="Model Selected",
              hue="Region")

plt.show()
plt.clf()
````

````python
# Create a pointplot and include the capsize in order to show caps and the error bars
sns.pointplot(data=df, y="Award_Amount",
              x="Model Selected",
              capsize=.1)

plt.show()
plt.clf()
````

````python
# Create a barplot with each Region shown as a different color
sns.barplot(data=df, y="Award_Amount",
            x="Model Selected",
            hue="Region")

plt.show()
plt.clf()
````

## Regression Plots

Since regression plots are a very important tool in the data scientist's toolbox, Seaborn has a robust API that supports sophisticated analysis of datasets.

In this lesson we will look at regression analysis of bicylce-sharing data from Washington, DC.

### Bicycle Dataset

* Aggregated bicycle sharing data in Washington DC
* Data includes :
    - Rental Amounts
    - Weather information
    - Calendar information
* Can we predict rental Amounts?

We can use Seaborn's regression plotting tools to evaluate the data from multiple perspectives and look for relationships between these numeric variables.

````python
# Bicycle Dataset
df = pd.read_csv('datasets/bike_share.csv')
````

### Plotting with regplot()

```regplot()``` function requires:
* Definiton of data
* x and y variables

````python
sns.regplot(data=df, x="temp",
            y="total_rentals", marker="+")
plt.show()
````

It looks like there is a relationship between temp and total bike rentals in a day. People tend to rent a bike when the weather is warm.

### Evaluating regression with residplot()

The residual plot is a very useful plot for understanding the appropratneness of a regression model. deally, the residual values in the plot should be plotted randomly across the horizontal line.

* A residual plot is useful for evaluating the fit of a model
* Seaborn supports through ```residplot``` function

In this example, the data looks like it might have a slight curve suggesting a nonlinear model might be appropriate.

````python
sns.residplot(data=df, x="temp", y="total_rentals")
plt.show()
````

### Polynomial Regression

If a value greater than 1 is passed to the **order** parameter of ```regplot()```, then Seaborn will attempt a polynomial fit using underlying NumPy functions.

* Seaborn supports polynomial regression using the **order** parameter

In this example, Seaborn computes a second order polynomial function for the relationship between temp and total rentals. It looks like rental start to trail off it the weather gets too warm.

````python
sns.regplot(data=df, x="temp",
            y="total_rentals", order=2)
plt.show()
````

#### Residplot with polynomial regression

* Residual plot can interpret the second order polynomial and plot the residual values.

In this example, the values are more randomly distributed, so a second order equation is likely more appropriate for this problem.

````python
sns.residplot(data=df, x="temp",
              y="total_rentals", order=2)
plt.show()
````

### Categorical Values

Seaborn also supports regression plots with categorical variables. It might be interesting to see how rentals change over the various months.

In this example, using **jitter** parameter makes it easier to see the individual distribution of the rental values for each month.

````python
sns.regplot(data=df, x="mnth", y="total_rentals",
            x_jitter=.1, order=2)
plt.show()
````

### Estimators

In some casesi even with the jitter, it may be difficult to see if there are any trends based on the value of the variable.

* In some cases, an ```x_estimator``` can be useful for highlighting trends

This simplified view shows a trend consistent with the seasons in Washington DC.

````python
# Import numpy
import numpy as np

sns.regplot(data=df, x="mnth", y="total_rentals",
            x_estimator=np.mean, order=2)
plt.show()
````

### Binning the Data

When there are continuous variables, it can be helpful to break them into different bins. In this case, we can divide the temp into four bins and Seaborn will take care of calculating those bins for us and plotting results.

* ```x_bins``` can be used to divide the data into discrete bins
* The regression line is still fit against all the data

This is much quicker than trying to use pandas or some other mechanism to create the bins

````python
sns.regplot(data=df, x="temp", y="total_rentals",
            x_bins=4)

plt.show()
````

## Matrix Plots

The final category of plots we will discuss in this section are **matrix** plots. The **heatmap** is the most common type of matrix plot and can be easily created by Seaborn.

Let's learn how to construct these plots in Seaborn and use them to look at our bicycle rental data.

````python
# Bicycle dataset
df = pd.read_csv('datasets/bike_share.csv')
````

### Getting data in the right format

* Seaborn's ```heatmap()``` function expects the data to be in *matrix*.
* pandas ```crosstab()``` function builds table to summarize the data by day and month

````python
pd.crosstab(df["mnth"], df["weekday"],
            values=df["total_rentals"],aggfunc="mean").round(0)
````

In this example, we also use the **aggfunc** argument to get the average number of rentals for each of the day and month combinations. This is very useful capability available in pandas and makes it easy to plot a heatmap once it is in this format.

### Build a heatmap

The *heatmap* translates the numerical values in the matrix into a color coded grid.

````python
sns.heatmap(pd.crosstab(df["mnth"], df["weekday"],
                        values=df["total_rentals"], aggfunc="mean"))

plt.show()
plt.clf()
````

In this example, the colors get lighter as the numbers increase. Using this approach, it is easy to see that the busiest days are Saturdays in June and September. The color code also illustrates that rental rates are higher in the warmer weather months from May through October.

#### Customize a heatmap

The display of the *heatmap* can be customized in multiple ways to present the most information as clearly as possible.

````python
df_crosstab = pd.crosstab(df["mnth"], df["weekday"],
                          values=df["total_rentals"], aggfunc="mean").astype(int)

sns.heatmap(df_crosstab, annot=True, fmt="d",
            cmap="YlGnBu", cbar=False, linewidths=.5)

plt.show()
plt.clf()
````

This example illustrates a number of customizations. 
* First, we use ```annot=True``` to turn on annotations in the individual cells. 
* The ```fmt``` option ensures that the results are displayed as integers. 
* We use a custom ```cmap``` of **Yellow Green Blue** to change the shading we use. 
* By setting ```cbar=False```, the color bar is not displayed. 
* Finally, passing a variable to ```linewidths``` puts small spacing between the cells so that the values are simpler to view. 
This example shows how changing a few parameters in Seaborn really changes the output of the plot.

#### Centering a heatmap

The final customization we will look at is centering the ```heatmap()``` color scheme.

````python
sns.heatmap(df_crosstab, annot=True, fmt='d',
            cmap='YlGnBu', cbar=True,
            center=df_crosstab.loc[9, 6])

plt.show()
plt.clf()
````

In this example, we center the colormap at the value stored for Sept and Saturday of 9 and 6 using 
the numeric indices for these values. The effect of this change is that overall color scheme is 
shifted towards yellows instead of blues.

### Plotting a correlation matrix

* Pandas ```corr``` function calculates correlations between columns in a dataframe
* The output can be converted to a heatmap with seaborn

This visualization can be useful to understand which variables you might want to further study using ```regplot()```.

````python
cols = ['total_rentals', 'temp', 'casual', 'hum', 'windspeed']
sns.heatmap(df[cols].corr(), cmap='YlGnBu')

plt.show()
plt.clf()
````
