## Changing plot Style and Color

By default, Seaborn plot are pleasing to look at, but there are several reasons you may want to change the appearance.

* Personal preference

* Improve readability

* Guide interpretation

### Changing the figure style

Seaborn has **five** figure styles which change the background and axes of the plot;

* **"white"**, **"dark"**, **"whitegrid"**, **"darkgrid"** and **"ticks"**

* To set one of these as the global style for all of your plots, use the ```sns.set_style()```


`````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Tips Dataset
tips = sns.load_dataset('tips')
tips.head()
`````

`````python
# Default figure style ("white")
sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            hue="sex")

# Show plot
plt.show()
`````

`````python
# Figure Style: "whitegrid"
sns.set_style("whitegrid")

sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            hue="sex")

# Show plot
plt.show()
`````

`````python
# Figure Style: "ticks"
sns.set_style("ticks")

sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            hue="sex")

# Show plot
plt.show()
`````

`````python
# Figure style: "dark"
sns.set_style("dark")

sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            hue="sex")

# Show plot
plt.show()
`````

`````python
# Figure style: "darkgrid"
sns.set_style("darkgrid")

sns.catplot(x="day",
            y="total_bill",
            data=tips,
            kind="bar",
            hue="sex")

# Show plot
plt.show()
`````

### Changing the palette

You can change the color of the main elements of the plot.

* Figure "palette" changes the color of the main elements of the plot

* use the ```sns.set_palette()``` function

* Use the preset palettes or create a custom palette.

`````python
# Default figure style
sns.set_style("white")

# Default Palette
sns.catplot(x="day",
            data=tips,
            kind="count")

# Show plot
plt.show()
`````

`````python
# Diverging palette
sns.set_palette("RdBu") # red/blue diverging palette

# Default Palette
sns.catplot(x="day",
            data=tips,
            kind="count")

# Show plot
plt.show()
`````

### Sequential Palettes

This group of palettes are single color(or two colors blended) moving from light to dark values.

`````python
# Sequential palette
sns.set_palette("Greys") # Gray single sequantial palette

# Default Palette
sns.catplot(x="day",
            data=tips,
            kind="count")

# Show plot
plt.show()
`````

### Custom Palettes

You can also create your own custom palettes by passing in a list of color names, or a list of hex color codes.

```python
custom_palette = ["red","green","orange","blue","yellow","purple"]

sns.set_palette(custom_palette)

```

**Or with hex color codes**

```python
custom_palette = ['#FBB4AE','#B3CDE3','#CCEBC5','#DECBE4','#FED9A6','#FFFFCC','#E5D8BD','#FDDAEC','#F2F2F2']

```

### Changing the Scale

* Figure "context" changes the scale of the plot elements and labels

* you can use ```sns.set_context()``` function

* Smallest to largest: "paper", "notebook", "talk", "poster"


`````python
# Default context: "paper"
sns.catplot(x="day",
            data=tips,
            kind="count")

# Show plot
plt.show()
`````

`````python
# Larger context: "talk"
sns.set_context("talk")

sns.catplot(x="day",
            data=tips,
            kind="count")

# Show plot
plt.show()
`````

## Adding titles and Labels: Part 1

We create data visualizations to communicate information, and we can't do that effectively without a clear title and informative axis labels.

### FacetGrid vs AxesSubplot objects

Seaborn's plot functions create two different types of objects:

* ```FacetGrids```and ```AxesSubplots```

To figure out which type of object we're working with, first assign the plot output to a variable then, return the **type()** function.

```python
g = sns.catplot(x="heigth", y="weight", data= df)
type(g)

# Output is 
'''
matplotlib.axes._subplots.AxesSubplot
'''
```

A **FacetGrid** consists of one or more AxesSubplots, which is how it supports subplots.

| Object Type | Plot Types | Characteristics |
| ------------|:----------:|----------------:|
| ```FacetGrid``` | ```relplot(), catplot()``` | Can Create subplots |
| ```AxesSubplot``` | ```scatterplot(), countplot()```, etc. | Only Creates a single plot |


### Adding Title to FacetGrid

Remember that **"catplot()"** enables subplots, so it returns a FacetGrid object. To adda title to a FacetGrid object;
* First assign that plot to a variable.

* Then you can call ```variable.fig.suptitle()``` function

* To adjust the height of the title, you can use **"y"** parameter. Default being 1.

`````python
# Dataset
gdp_data = pd.read_csv("datasets\countries-of-the-world.csv")

gdp_data = gdp_data.astype({"Region": str}, errors="raise")
gdp_data["Birthrate"] = pd.to_numeric(gdp_data["Birthrate"], errors="coerce")

gdp_data.dtypes
gdp_data["Birthrate"].isnull()


# Create the box plot
g = sns.catplot(x="Region",
                y="Birthrate",
                data=gdp_data,
                kind="box")

# Adding a Title and adjusting its position
g.fig.suptitle("New Title",
               y=1.03)

# Show plot
plt.show(g)
`````

## Adding titles and Labels: Part 2

We learned how to add a title to a FacetGrid object using **g.fig.suptitle()**

To add a title to an AxesSubplot object;

* use **g.set_title()**

### Titles for subplots

What happens if the figure has subplots?

To alter the subplot titles, use **g.set_titles()** to set the titles for each AxesSubplot.

If you want to use the variable name in the title, you can use **{col name}**

**FacetGrid**

```python
g = sns.catplot(x="Region",
                y= "Birthrate",
                data=gdp_data,
                kind="box")

g.fig.suptitle("New Title",
                y=1.03)

```

**AxesSubplot**

```python
g = sns.boxplot(x="Region",
                y="Birthrate",
                data=gdp_data)

g.set_title("New Title",
            y=1.03)                
```

`````python
gdp_data.dtypes
`````

`````python
# Titles for subplots
g = sns.catplot(x="Country",
                y="GDP ($ per capita)",
                data=gdp_data,
                kind="point",
                col="Region")

# Adding title to FacetGrid
g.fig.suptitle("New Title",
               y=1.03)

# Adding title to AxesSubplot
g.set_titles("This is {col_name}")

# show plot
plt.show()
`````

### Adding axis labels

To add axis labels, assign the plot to a variable and then call the **set** function.

Set the parameters **"x label"** and **"y label"**

This works with both FacetGrid and AxesSubplot objects.

#### Rotating x-axis tick labels

To do this, we don't call a function on the plot object itself. Instead, after we create the plot, we call the matplotlib function **plt.xticks** and set **rotation** equal to 90 degrees. This works with both FacetGrid and AxesSubplot objects.

`````python
# Titles for subplots
g = sns.catplot(x="Country",
                y="GDP ($ per capita)",
                data=gdp_data,
                kind="point",
                col="Region")

# Adding title to FacetGrid
g.fig.suptitle("New Title",
               y=1.03)

# Adding title to AxesSubplot
g.set_titles("This is {col_name}")

# Adding axis labels
g.set(xlabel="New X Label",
      ylabel="New Y Label")

# Rotating the x-axis tick labels
plt.xticks(rotation=90)

# show plot
plt.show()
`````

## Exercise III

### Adding a title and axis labels

Let's continue to look at the miles per gallon dataset. 
This time we'll create a line plot to answer the question: How does the average miles per gallon achieved by cars change over time for each of the three places of origin? 
To improve the readability of this plot, we'll add a title and more informative axis labels.


`````python
# Creating a mpg_mean column
mpg.loc[:, "mpg_mean"] = mpg.groupby("mpg")["mpg"].transform("mean")

# Default color palette
custom_palette = ["Red", "Green", "Blue"]
sns.set_palette(custom_palette)

# Increase Scale
sns.set_context("notebook")

# Create a line plot
g = sns.lineplot(x="model_year",
                 y="mpg_mean",
                 data=mpg,
                 hue="origin",
                 ci=None)

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year",
      ylabel="Average MPG")
# Show plot
plt.show()
`````

### Rotating x-tick labels

In this exercise, we'll continue looking at the miles per gallon dataset. 
In the code provided, we create a point plot that displays the average acceleration for cars in each of the three places of origin. 
Note that the ```"acceleration"``` variable is the time to accelerate from 0 to 60 miles per hour,
in seconds. Higher values indicate slower acceleration.

`````python
# Create point plot
sns.catplot(x="origin",
            y="acceleration",
            data=mpg,
            kind="point",
            join=False,
            capsize=.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
`````

### Relational vs. Categorical Plots

After we've imported the appropriate libraries, the next thing to do is to choose what type of plot we want to create.

| Relational Plots | Categorical Plots |
| ----------------:| -----------------:|
| Shows the relationship between two quantitative variables | Shows the distrubition of a quantitative variable within categories defined by a categorical variable |
| Eg: Scatter plots, line plots | Eg: bar plots, count plots, box plots, point plots |
| ```sns.relplot(x="x_variable_name",y="y_variable_name",data=pandas_df,kind="scatter")``` | ```sns.catplot(x="x_variable_name", y="y_variable_name", data=pandas_df, kind="bar")``` |
              

### Adding a third variable (hue)

If we want to add a third dimension to our plots, we can do this in of two ways. Setting the **"hue"** parameter to a variable name will create a single plot but will show subgroups that are different colors based on that variable's values.

### Adding a third variable (row/col)

We can use **"relplot()"** and **"catplot()"'s** and **"row"** and **"col"** parameters to graph each subgroup on a separate subplot in the figure.(eg: Smoker yes/no in tips dataset)

### Customization

* Change the background : ```sns.set_style()```

* Change the main element colors : ```sns.set_palette()```

* Change the scale : ```sns.set_context()```

### Adding a Title

| AxesSubplot | scatterplot(), countplot(), etc. | g.set_title() |
